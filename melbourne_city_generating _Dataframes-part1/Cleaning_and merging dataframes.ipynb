# extract the distance between furniture location and sensor location

```python

D = {'distance_sensor': [], 'sensor_id':[] }
i=1;
for furniture_ind in furniture.index:
    min_dist = 9999
    sensor_id = 0
    for sensor_ind in pedestrian_sensor_data.index:
        #get the furniture and sensor location with spaces and brackets stripped out
        furniture_loc = furniture['location'][furniture_ind].strip().replace(" ","")
        pedestrian_loc = pedestrian_sensor_data['location'][sensor_ind].strip().replace(" ","").replace(")","").replace("(","")
        #print("furniture loc:" + furniture_loc + " sensor loc: " + pedestrian_loc)
        
        #partition the furniture and sensor locaiton to get latitude and longitude separately
        furn_part = furniture_loc.partition(',')
        sensor_part = pedestrian_loc.partition(',') 
        
        #populate latitude and longitude into loc as tuples
        loc1=(float(furn_part[0]), float(furn_part[2]))
        loc2=(float(sensor_part[0]), float(sensor_part[2]))

        #print("furn: " + str(loc1) + "sensor: " + str(loc2))
        distance = hs.haversine(loc1, loc2)
        if distance < min_dist:
            min_dist = distance
            sensor_id = pedestrian_sensor_data['sensor_id'][sensor_ind]
            
        #testing
        #print("sensor id: "+str(pedestrian_sensor_data['sensor_id'][sensor_ind])+" dist:"+str(distance)+str(loc1)+str(loc2))
        
        #dst_list.append(str(distance) + "," + str(pedestrian_sensor_data['sensor_id'][sensor_ind]))
        #distance_sensor=[min(dst_list),dst_list.index(min(dst_list))]
        
        ####################################################################
        #end of inner (sensor) Loop
        ####################################################################
        
    #break

#print(min_dist, sensor_id)

        
#print(dst_list)
#print(dst_list.index(min(dst_list)))
#print("sensor id:" + str(pedestrian_sensor_data['sensor_id'][dst_list.index(min(dst_list))]))
    D['distance_sensor'].append(min_dist)
    D['sensor_id'].append(sensor_id)
   
```

#  Generating a new dataframe with distance between furniture and sensor location and sensor id

```python
output = pd.DataFrame(D)
output
furniture['distance_sensor']=output['distance_sensor']
furniture['sensor_id']=output['sensor_id']
```
